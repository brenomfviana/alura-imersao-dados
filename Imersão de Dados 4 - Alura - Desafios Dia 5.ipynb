```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

pd.set_option("display.precision", 2)
pd.set_option("display.float_format", lambda x: "%.2f" % x)
sns.set()

data_censo = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/datasets/dados_vendas_censo.csv")
data_censo = data_censo[["Rua", "Bairro", "Cidade", "Metragem", "Quartos", "Banheiros", "Vagas",
       "Valor_anuncio", "Valor_mm", "Valor_m2", "latitude", "longitude", "Situacao_setor", "V001",
       "V002", "V003", "V004", "V005", "V006", "V007", "V008", "V009", "V010",
       "V011", "V012"]].dropna()
data_censo
```

## 2 - Test other regression models.

```python
from sklearn.model_selection import train_test_split

X = data_censo[["Metragem"]]
Y = data_censo["Valor_anuncio"]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=54)
```

```python
Y_test
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
Y_predict_test = lr.predict(X_test)
Y_predict_test
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
mean_absolute_error(Y_test, Y_predict_test)
```

### SVM Regression

```python
from sklearn import svm

regr = svm.SVR()
```

```python
regr.fit(X, Y)
```

```python
Y_predict_test = regr.predict(X_test)
Y_predict_test
```

```python
mean_absolute_error(Y_test, Y_predict_test)
```

### Decision Tree Regression

```python
from sklearn import tree

clf = tree.DecisionTreeRegressor()
```

```python
clf = clf.fit(X, Y)
```

```python
Y_predict_test = clf.predict(X_test)
Y_predict_test
```

```python
mean_absolute_error(Y_test, Y_predict_test)
```

## 3 - Create a baseline model with the original features.

```python
mean_absolute_error(Y_test, Y_predict_test)
```

```python
regr = svm.SVR()
regr.fit(X, Y)
Y_predict_test = regr.predict(X_test)
Y_predict_test
```

```python
clf = tree.DecisionTreeRegressor()
clf = clf.fit(X, Y)
Y_predict_test = clf.predict(X_test)
Y_predict_test
```

```python
mean_absolute_error(Y_test, Y_predict_test)
```
