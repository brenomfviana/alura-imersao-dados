## 2. Find two ways of selecting data by district

```python
# Get the district names
districts = data["Bairro"].drop_duplicates()
districts
```

```python
# 1
district = data["Bairro"] == districts[0]
data[district]
```

```python
# 2
data.where(data["Bairro"] == districts[0]).dropna(subset=["Bairro"])
```

```python
# 3
district = districts[0]
data.query('Bairro == @district')
```

## 3. Explore some graphics in the documentation and apply them to the other columns of the database, try to observe something

```python
top10districts = data["Bairro"].value_counts().head(10)
top10districts
```

Only three districts have more than 300 properties available: Alto de Pinheiros, Jardim Guedala, Jardim Paulista

```python
top10districts.plot.bar()
```

Get all properties of the top 10 districts

```python
properties_top10districts = data["Bairro"].value_counts().head(10)
properties_top10districts = data[data["Bairro"].isin(top10districts.keys())]
properties_top10districts
```

Almost half of the properties have 4 rooms and almost 30% have 3 rooms

```python
properties_top10districts["Quartos"] \
  .value_counts() \
  .sort_index() \
  .plot.pie(subplots=True, autopct="%.2f")
```

More than half properties have at least 5 bathrooms

```python
properties_top10districts["Banheiros"] \
  .value_counts() \
  .sort_index() \
  .plot.pie(subplots=True, autopct="%.2f")
```

Most properties have 3 parking spaces

```python
properties_top10districts["Vagas"].plot.hist(bins=50)
```

There is a huge property available

```python
properties_top10districts["Metragem"].plot.box()
```

Remove the biggest property

```python
properties_top10districts = properties_top10districts.where(properties_top10districts["Metragem"] < 700000)
properties_top10districts
```

Half of the properties are around 500 square meters and are less than 1000 square meters

```python
properties_top10districts["Metragem"].plot.box()
```
